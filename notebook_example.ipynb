```python
import pickle
import pandas as pd
import pdpfunct as fclust
```

```python
root_sentiero = "../../joschi_model_original_data/wine"
dfTest = pd.read_csv(root_sentiero+"/test.csv")
name = "rf"
fileobj = open( root_sentiero+"/"+name+".pkl", "rb" )
pickle_model = pickle.load( fileobj )
featureIndex = 0
numSamples = 100
clustNumber = 10
scale = pickle_model["scale"]
shift = pickle_model["shift"]
model = pickle_model["model"]
labels_name = model.classes_
```

```python
fclust.plot_pdp(dfTest,
             model,
             featureIndex,
             numSamples,
             clustNumber,
             scale,
             shift,
             labels_name,
             thresh = 0.5,
             LBKeogh = False,
             spaghetti = True, 
             ChunksYes = False)
```
